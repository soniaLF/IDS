## Funzione per ridurre l'uso della memoria

Funzione che cerca di utilizzare i tipi di dato piÃ¹ efficienti in termini di memoria per le colonne numeriche, mantenendo al contempo la precisione necessaria dei dati. 


```python
def reduce_memory_usage(df):
    for col in df.columns:
        col_type = df[col].dtype
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                else:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    return df
```

## Regolarizzazione min max

```python
# Carica il dizionario min-max
with open('../archive/min_max_values.json', 'r') as f:
    min_max_values = json.load(f)

# Specifica le feature da regolarizzare nel test set
features = list(min_max_values.keys())


# Filtro il DataFrame per tenere solo le colonne di interesse
df_features_test = test2[features].copy()

# Normalizzazione manuale basata sui valori min e max
def normalize_column(col, min_val, max_val):
    return (col - min_val) / (max_val - min_val)

# Applica la normalizzazione a ciascuna feature
for feature in features:
    min_val = min_max_values[feature]['min']
    max_val = min_max_values[feature]['max']
    df_features_test.loc[feature] = normalize_column(df_features_test[feature], min_val, max_val)

# Aggiungi le colonne scalate al DataFrame originale
test2[features] = df_features_test

# Mostra il DataFrame trasformato
print(test2.head())
```
