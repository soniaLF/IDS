# Funzione per ridurre l'uso della memoria

Funzione che cerca di utilizzare i tipi di dato più efficienti in termini di memoria per le colonne numeriche, mantenendo al contempo la precisione necessaria dei dati. 


```python
def reduce_memory_usage(df):
    for col in df.columns:
        col_type = df[col].dtype
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                else:
                    df[col] = df[col].astype(np.int64)
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
    return df
```

# Caricamento del dataset

```python
file_path = '../archive/dataset14-15-16-21-22-23-28-01-02-20.csv'
df = pd.read_csv(file_path)
df = reduce_memory_usage(df)
```

# Controllo duplicati

```python
# Controlla se ci sono righe duplicate
duplicates1 = df.duplicated().sum()

# Visualizza il numero di righe duplicate
print(f"Numero di righe duplicate: {duplicates1}")
```

Controllo quanti duplicati siano benigni prima di eliminarli 

```python
# Trova i duplicati (considerando tutte le colonne)
duplicates1 = df[df.duplicated()]

# Conta il numero totale di duplicati
total_duplicates = len(duplicates1)

# Filtra i duplicati per quelli con 'Label' == 'Benign'
benign_duplicates = duplicates1[duplicates1['Label'] == 'Benign']

# Conta il numero di duplicati con 'Label' == 'Benign'
benign_duplicates_count = len(benign_duplicates)

# Stampa i risultati
print(f"Numero totale di duplicati: {total_duplicates}")
print(f"Numero di duplicati con 'Label' == 'Benign': {benign_duplicates_count}")
```

Sono tutti benigni ha senso eliminarli anche per bilanciare il dataset

```python
# Elimina i duplicati
df_cleaned = df.drop_duplicates()

# Verifica il numero di righe dopo la rimozione dei duplicati
duplicates2 = df_cleaned.duplicated().sum()

# Visualizza il numero di righe duplicate dopo la rimozione
print(f"Numero di righe duplicate dopo la rimozione: {duplicates2}")
```

# Matrice di correlazione 

```python

# Calcola la matrice di correlazione
corr = df.corr()

# Configura la dimensione della figura
plt.figure(figsize=(40, 40))

# Crea l'heatmap della matrice di correlazione
sns.heatmap(corr, annot=True, cmap='RdBu', vmin=-1, vmax=1, square=True, fmt='.2f')

# Mostra la figura
plt.show()

# Salva la figura in un file 
plt.savefig('heatmap.png', bbox_inches='tight')
```

# Eliminazione feature altamente correlate (>0,9)

```python
# Separa la colonna 'Label' (o qualsiasi colonna di tipo object che vuoi mantenere)
label_col = df['Label']
df_numeric = df.drop(columns=['Label'])

# Calcola la matrice di correlazione
corr = df_numeric.corr().abs() 

# Imposta una soglia di correlazione
threshold = 0.9 

# Trova le coppie di feature con correlazione sopra la soglia
# Mask per ottenere solo l'upper triangle della matrice di correlazione
mask = (corr > threshold) & (corr < 1.0)

# Trova gli indici delle colonne da rimuovere
to_drop = set()
for i in range(len(mask.columns)):
    for j in range(i):
        if mask.iloc[i, j]:
            colname = mask.columns[i]
            to_drop.add(colname)

# Rimuovi le feature molto correlate
df_reduced = df.drop(columns=to_drop)

# Riaggiungi la colonna 'Label'
df_reduced['Label'] = label_col


# Visualizza le feature rimaste
print("Feature rimosse:", to_drop)
print("Feature rimaste:", df_reduced.columns.tolist())
```
